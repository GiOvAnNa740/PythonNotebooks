### General information

    Methods require parenthesis, attributes do not

In order to 'save' the filtered date, we can use the argument "inplace=True" or method the mathod .inplace() so that the the filtered 'view' will overlap the data and rewrite it. 

Other option would be to reatribute the filtered data to another variable or to itself

.apply() method -> Allows you to invoke a function on every value on a series

.map() method -> Simmilar to the apply method, but it accepts an associative object (such as a dictionary or another pandas series) instead of a function

df.insert(loc=2, column='', value='') -> Will insert the specified columns on the defined position without overwriting the existing column

df.dropna() -> Will completely remove any rows with missing values

    Arguments
    
how='any' -> default

how='all' -> will remove only lines were all values are missing

subset=['column_name'] -> Will remove values only on the specified column(s)

```python
# will return True or False telling if there are any missing values on the dataframe

nba.isnull().values.any()
```

```python
nba = nba.dropna(how='all')
```

df.fillna('example') -> Will fill the missing values with the passed argument

*Can be set to a specific column or a specifi group of columns

```python
nba['Salary'] = nba['Salary'].fillna(0) # filling the missing values with zeros
```

df.hasnans['columns'] -> Returns a boolean if the data has missing values or not

df['column'].astype(int) -> converts the data to the specified format

```python
nba['Salary'] = nba['Salary'].astype(int)
```

df['column'].nunique() -> Retuns the quantity of unique values

```python
nba['Team'].nunique() #will return only th number of unique values
```

```python
nba['Team'].unique() #Returns an array with a list of the unique values
```

### Sorting and rearranging the dataFrame

df.sort_values(by='column_name') -> The data will be sorted based on the specified column or list of columns (on a list, python will follow the set order of columns)

    Arguments (in case of multiple columns, we can set multiple arguments in reference of each column)

na_position='last'/'first' -> lets you define where the NaN values will be sorted

ascending=True/False -> Set ascending or descending order

```python
nba
```

```python
nba.sort_values(by=['Team','Name'], ascending=[True, False]) # we can sort each column on a different way
```

df.index() -> The data will be sorted based on the indexes

```python
nba.sort_index() # good way to 'reset' the values to the original order
```

df['column_name'].rank() -> will rank the values of the columns in numbers from 1 to N

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending=False).astype(int) # Creating a rank for the salaries on a new column

nba.sort_values('Salary',ascending=False) # sorting the new df to show the ranked salaries
```

### Filtering by comparison

-   ==
-   !=
-   '>'
-   '<'
-   '>='
-   '<='

```python
# this will return a Series of booleans
employees['Gender'] =='Female'
```

```python
# this will return our df filtered based on the specified condition
employees[employees['Gender'] =='Female']
```

    We may also apply the filter to another variable

```python
mask = employees['Team'] !='Finance' # any values that is not 'Finance'
employees[mask]
```

    On a boolean set of values python will assume we are willing to filter onle the values that are True

```python
employees[employees['Senior Management']]
```

Other operators may be applied

```python
employees[employees['Salary'] >=100000]
```

### Filtering with more conditions

-   AND => &
-   OR => |

```python
female=employees['Gender']=='Female'
manager=employees['Senior Management'] # if 'Senior Management' == True
team=employees['Team']=='Engineering'
```

AND

```python
employees[female & manager]
```

OR

```python
employees[female | manager]
```

Both -> The partenthesis are necessary to define the priority

```python
employees[female & (team | manager)]
```

.isin() Method

```python
mask = employees['Team'].isin(['Product','Finance','Engineering']) # this will return a Series of booleans

employees[mask]
```

.between() Method

```python
salary = employees['Salary'].between(50000,60000)

employees[salary]
```

.duplicate Method

```python
employees['First Name'].duplicated()
```

```python
~employees['First Name'].duplicated() # the ~ symbol means 'not', so here we're checking for all that are NOT duplicates, so unique values
```

```python
duplicates = employees['First Name'].duplicated(keep='last') # keep= lets you define which of the repeated values will be kept, by default is the 'first',
#it can also be set to False to remove ALL duplicates
employees[duplicates]
```

.drop_duplicates method

```python
employees.drop_duplicates(subset=['First Name'], keep='first')
```

```python
employees.drop_duplicates(subset=['First Name','Team'], keep=False,  # will only remove rows were the name AND Team are the same
                          inplace=True) # will overwrite the dataframe, it is 'False' by default

employees
```
