```python
import pandas as pd
import numpy as np
```

## Changing options and setting in pandas

```python
data = np.random.randint(0,100,[1000,50]) # random numbers from 0 to 100 in a matrix of 1000 by 50

df=pd.DataFrame(data)

df
```

### Changing some parameters

### .options.display

```python
pd.options.display.max_rows #original diaplay value
```

```python
pd.options.display.max_rows = 5
df
```

```python
pd.options.display.max_columns
```

```python
pd.options.display.max_columns = 10
df
```

### .reset_option()

```python
pd.reset_option('max_rows')

pd.options.display.max_rows
```

### .describe_option()

loads the documentation

```python
pd.describe_option('max_columns')
```
