### Using a URL as input

```python
url = 'https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv'

baby_names = pd.read_csv(url)
baby_names
```

    .to_list()

```python
baby_names["Child's First Name"].to_list()
```

```python
baby_names["Child's First Name"]
```

```python
', '.join(baby_names["Child's First Name"] # will separate each item on column "Child's First Name" with a comma ', ' and join them in a single String
        .drop_duplicates() #df method
        .sort_values() #df method / sort alphabeticaly
        ).title() #string method
```

### Exporting csv data

```python
baby_names.to_csv('db/NYC_baby_names.csv', index=True,# by default an index column will be included, may be set to false
                 encoding='utf-8') #providing an encoding to avoid special unwritable characters  

# you may also provide only somespecific columns to save
#baby_names.to_csv('db/NYC_baby_names.csv', index=True, columns=['Gender',"Child's First Name"]) 
```

### Exporting excel data

```python
girls=baby_names[baby_names['Gender'] == 'FEMALE']
boys=baby_names[baby_names['Gender'] == 'MALE']
```

    Creating an empty excel file

```python
excel_file=pd.ExcelWriter('db/baby_names.xlsx')
```

    Creating the sheets

```python
girls.to_excel(excel_file, sheet_name='Girls', index=False)
```

```python
boys.to_excel(excel_file, sheet_name='Boys', index=False)#you may add a columns parameter to specify columns
```

    Joining all on the escel file

```python
excel_file.save()
```
