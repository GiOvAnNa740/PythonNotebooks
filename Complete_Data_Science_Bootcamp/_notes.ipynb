## Combinatorics -> Combinations of objects of a specific finite set

Restriction can be applied such as:
Repetition, Order and others

The elements of Combinatorics are:
-Permutation
-Variations
-Combinations


-> $n$ = Total number of elements

-> $p$ = Number of positions to be filled


===

    #Permutation -> used when all elements need to be arranged on the space, when each element will be picket one at a time

$P_{n} = n!$

Obs: 

-> Negative numbers do not have a factorial

-> $0! = 1$

-> $n! = (n-1)! * n$

-> $(n+1)! = n! * (n+1)$


===

    #Variation -> used when not all elements will be used at the same time (repeting or not), when each element will be picket one at a time


=> Variation formula with repetition

$\overline{V}_{p}^{n} = n^{p}$


=> Variation formula without repetition

${V}_{p}^{n} = \frac{n!}{(n-p)!}$


===

    #Combinations -> used when a pair or group of elements will be picked together for each choice, for this, order is irrelevant

=> Combination formula with repetition

$\overline{C}_{p}^{n} = \frac{\overline{V}_{p}^{n}}{P_{p}} = \frac{(n+p-1)!}{p!(n-1)!}$

=> Combination formula without repetition

${C}_{p}^{n} = \frac{V_{p}^{n}}{P_{p}} = \frac{n!}{p!(n-p)!}$ 

-> When we have separate sample spaces, the combination is solved dimply by multiplying each possibility 

(Example: Possible combinatios for lunch including 3 main, 4 beverages, and 2 desserts would simply be: $3 * 4 * 2$)

!Combination results are always simetric

${C}_{p}^{n} = {C}^{n}_{n-p}$

    #Conditional Probability

-> Independent events

$P(A) = P(A|B)$

-> Dependent events

$P(A|B) = \frac{P(A \cap B)}{P(B)}    for P(B) > 0$

    #Total Law of probability

$P(A) = P(A|B_{1})*P(B_{1}) + P(A|B_{2})*P(B_{2})...+ P(A|B_{n})*P(B_{n})$


    #Aditive Law

$P(A \cup B) = P(A)+P(B) - P(A \cap B)$


#Multiplication Law

$P(A \cap B) = P(A)*P(B)$ Given that B depends on A

    #Bayes' Law/Rule/Theorem

$P(A|B)=\frac{P(A \cap B)}{P(B)}$ and $P(A \cap B) = P(B|A)*P(A)$



# Probability Distributions

$p(Y=y)$ or $P(y)$

-> $Y$ : Actual outcome of an event

-> $y$ : One of the possible outcomes

-> Mean: $\mu$

-> Variance: $\sigma^{2}=E(Y^{2})-\mu^{2}$

-> Sample/Standart Deviation: $\sqrt{\sigma^{2}}=\sigma$

-> Sample population: $\overline{x}$

-> Sample Variation: $S$

-> Degres of Freedom: $k$


# Discrete Distributions

    #Uniform distributions: all outcomes have the same propability

->Offers no relevat information

-> Values are specifically defined

    #Bernoulli (Bern) Distribution: Events with one trial and only two possible outcomes

    #Binomial Distribution: Conducting a single experiment several times where each event has only two possible outcomes

-> Probability Function:

$C^{n}_{y}$

$p(y)=(^{n}_{y})*p^{y}*(1-p)^{n-y}$

    #Poisson Distribution: How unusual an event frequency is for a given interval

$P(Y)=\frac{\lambda^{y}e^{-\lambda}}{y!}$

$e$ = Euler's number

# Continuos Distribution: The probability will be a curve. Infinitely many consecutive outcomes

->Ranges to and from infinite within a range, graphs being curves were all infinite values between numbers are contained

$P(x>X) = P(x\geqslant X)$

$P(x<X) = P(x\leqslant X)$

    #Normal Distribution: Often observed in nature and outliers are extremelly rare

-> Usually requires a lot of data

-> Bell shaped, short tails

    #Student's T distribution: Small sample approximation of a Normal distribution. It accomodates extreme values better

-> Bell shaped, larger tails

-> also need the Degrees of Freedom $k$ to be defined

    #Chi-Squared Distribution: Asymmetric and only consists of non-negative values

-> Used in hypothesys testing

-> Asymetric graph starting from zero, higher on the left side

-> Expected value $E(x)=k$, the Variance $Var(X)=2k$

    #Exponential Distribution: Used in Events that are rapidly changing early on

-> PDF approaches 0 while the CDF approaches 1

    #Logistic Distribution: Useful in forecast analisys when determining a key point for a successful outcome 

-> PDF graph bell-shaped, simmilar to the Normal Distribution

-> CDF graph grows to the right side, driving drastically up after the mean point, approaching 1

!No prediction is ever a true certainty

PDF -> Graph of Probability Density Function

PDC -> Probability Distribution Curve

CDF -> Cumulative Distribution Function 

Standardizing -> Transforming the graph so that the expected value $E(x)=0$, the Variance $Var(X)=1$, the Mean $\mu=0$ and Standart Deviation $\sigma=1$
