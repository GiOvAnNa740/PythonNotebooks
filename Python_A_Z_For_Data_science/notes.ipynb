# Seaborn: Data visualizer

Seaborn Gallery -> useful to find graph styles
https://seaborn.pydata.org/examples/index.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline

#ignoring warnings
import warnings
warnings.filterwarnings('ignore') 

#increasing the size of the graph
plt.rcParams['figure.figsize'] = 8,4 
```

```python
#importing a Dataframe

stats = pd.read_csv('df\P4-Demographic-Data.csv')

stats.columns
```

```python
#Distribution

vis1 = sns.distplot(stats['Internet users'],bins=30)
```

```python
#Boxplot

vis2 = sns.boxplot(data=stats, x='Income Group', y='Birth rate')
```

```python
#Linear Model with scatterPlots

vis3 = sns.lmplot(data=stats, x='Internet users', y='Birth rate', aspect=1.5 ) # 'aspect' changes the size ratio
```

```python
#Scatter model with colors

vis4 = sns.lmplot(data=stats, x='Internet users', y='Birth rate',
    fit_reg=False, hue='Income Group')
```

Keyword Arguments

-> A way to include aditional arguments to a library

-> It is necessary to now the name of the element and the property to be altered

```python
#Scatter model with aditional arguments

vis4 = sns.lmplot(data=stats, x='Internet users', y='Birth rate',
    fit_reg=False, hue='Income Group',
    scatter_kws={'s':80}) #changing the size ('s') of the scattered points
```

```python
movies.Genre

genreLabels = list()

genreLabels = movies.Genre.cat.categories

genreLabels
```

```python
#Overlaped graph

plt.hist(movies[movies.Genre == 'Action'].BudgetMillions, bins=10)
plt.hist(movies[movies.Genre == 'Adventure'].BudgetMillions, bins=10) 
plt.hist(movies[movies.Genre == 'Comedy'].BudgetMillions, bins=10) 
plt.hist(movies[movies.Genre == 'Drama'].BudgetMillions, bins=10) 
plt.hist(movies[movies.Genre == 'Horror'].BudgetMillions, bins=10) 
plt.hist(movies[movies.Genre == 'Romance'].BudgetMillions, bins=10) 
plt.hist(movies[movies.Genre == 'Thriller'].BudgetMillions, bins=10) 

plt.show()
```

```python
#Grouped graph

plt.hist([movies[movies.Genre == 'Action'].BudgetMillions,
        movies[movies.Genre == 'Adventure'].BudgetMillions, 
        movies[movies.Genre == 'Comedy'].BudgetMillions,
        movies[movies.Genre == 'Drama'].BudgetMillions,
        movies[movies.Genre == 'Horror'].BudgetMillions,
        movies[movies.Genre == 'Romance'].BudgetMillions,
        movies[movies.Genre == 'Thriller'].BudgetMillions],
        bins=10, label=genreLabels)

plt.legend()
plt.show()
```

```python
#Stacked graph

plt.hist([movies[movies.Genre == 'Action'].BudgetMillions,
        movies[movies.Genre == 'Adventure'].BudgetMillions, 
        movies[movies.Genre == 'Comedy'].BudgetMillions,
        movies[movies.Genre == 'Drama'].BudgetMillions,
        movies[movies.Genre == 'Horror'].BudgetMillions,
        movies[movies.Genre == 'Romance'].BudgetMillions,
        movies[movies.Genre == 'Thriller'].BudgetMillions],
        bins=20, stacked=True,label=genreLabels)

plt.legend()
plt.show()
```

Better way to include all genres:

```python
for gen in movies.Genre.cat.categories: # for this function to work the data type must be set as 'category', "movies.Genre" will return the full Genre column and not only the unique labels
    print(gen)
```

```python
list1 = list() # or list1 = []
labels = list()

for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    labels.append(gen) #creating the lables list

plt.hist(list1, bins=20, stacked=True,
         label=labels) #including the labels to the histogram

plt.legend() #showing the labels
```

#Subplots

Combining visualizations in matrices

```python
f, axis = plt.subplots(1,2, figsize=(12,6), sharex=True, sharey=True) # (rows, columns) | 'sharex/sharey' makes all graphs share the same x and y limits

# f, ax = returns a tuple, 'f' is for the figure, 'ax' is for the axis position

k1 = sns.kdeplot([movies.AudienceRating, movies.BudgetMillions],shade=True, ax=axis[0]) 
k2 = sns.kdeplot([movies.RottenTomatoesRating, movies.BudgetMillions],shade=True, ax=axis[1]) 

k1.set(xlim=(-50, 300))


# [1] is only valid for 1 dimension, as soon as the matrix has more than 1 dimention it is necessary to specify lines and columns. Ex: [0,1]
```
