# Seaborn: Data visualizer

Seaborn Gallery -> useful to find graph styles
https://seaborn.pydata.org/examples/index.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline

#ignoring warnings
import warnings
warnings.filterwarnings('ignore') 

#increasing the size of the graph
plt.rcParams['figure.figsize'] = 8,4 
```

```python
#importing a Dataframe

stats = pd.read_csv('df\P4-Demographic-Data.csv')

stats.columns
```

```python
#Distribution

vis1 = sns.distplot(stats['Internet users'],bins=30)
```

```python
#Boxplot

vis2 = sns.boxplot(data=stats, x='Income Group', y='Birth rate')
```

```python
#Linear Model

vis3 = sns.lmplot(data=stats, x='Internet users', y='Birth rate', aspect=1.5 ) # 'aspect' changes the size ratio
```

```python
#Linear model with colors

vis4 = sns.lmplot(data=stats, x='Internet users', y='Birth rate',
    fit_reg=False, hue='Income Group')
```

Keyword Arguments

-> A way to include aditional arguments to a library

-> It is necessary to now the name of the element and the property to be altered

```python
#Linear model with aditional arguments

vis4 = sns.lmplot(data=stats, x='Internet users', y='Birth rate',
    fit_reg=False, hue='Income Group',
    scatter_kws={'s':80}) #changing the size ('s') of the scattered points
```

#Changing the data type

For the 'Year' it does not make sense to be treated as an integer

```python
movies.Film = movies.Film.astype('category')  # without the atribution the data will not be changes on the DataFrame
movies.Genre = movies.Genre.astype('category')

movies.ReleaseYear = movies.ReleaseYear.astype('category')
```

#JointPlots

```python
jplot = sns.jointplot(data=movies, x='RottenTomatoesRating', y='AudienceRating')
```

```python
#changing the data plot type

#Hex highlights the zones with the most occurencies

jplotHex = sns.jointplot(data=movies, x='RottenTomatoesRating', y='AudienceRating',
    kind='hex')
```
