# Seaborn and Pyplot: Data visualizers

Seaborn Gallery -> useful to find graph styles
https://seaborn.pydata.org/examples/index.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline

#ignoring warnings
import warnings
warnings.filterwarnings('ignore') 

#increasing the size of the graph
plt.rcParams['figure.figsize'] = 8,4 
```

```python
#importing a Dataframe

stats = pd.read_csv('df\P4-Demographic-Data.csv')

stats.columns
```

```python
#Distribution

vis1 = sns.distplot(stats['Internet users'],bins=30)
```

```python
#Boxplot

vis2 = sns.boxplot(data=stats, x='Income Group', y='Birth rate')
```

```python
#Linear Model with scatterPlots

vis3 = sns.lmplot(data=stats, x='Internet users', y='Birth rate', aspect=1.5 ) # 'aspect' changes the size ratio
```

```python
#Scatter model with colors

vis4 = sns.lmplot(data=stats, x='Internet users', y='Birth rate',
    fit_reg=False, hue='Income Group')
```

Keyword Arguments

-> A way to include aditional arguments to a library

-> It is necessary to now the name of the element and the property to be altered

```python
#Scatter model with aditional arguments

vis4 = sns.lmplot(data=stats, x='Internet users', y='Birth rate',
    fit_reg=False, hue='Income Group',
    scatter_kws={'s':80}) #changing the size ('s') of the scattered points
```

#Subplots

Combining visualizations in matrices

```python
f, axis = plt.subplots(1,2, figsize=(12,6), sharex=True, sharey=True) # (rows, columns) | 'sharex/sharey' makes all graphs share the same x and y limits

# f, ax = returns a tuple, 'f' is for the figure, 'ax' is for the axis position

k1 = sns.kdeplot([movies.AudienceRating, movies.BudgetMillions],shade=True, ax=axis[0]) 
k2 = sns.kdeplot([movies.RottenTomatoesRating, movies.BudgetMillions],shade=True, ax=axis[1]) 

k1.set(xlim=(-50, 300)) #setting the limits to the X axis


# [1] is only valid for 1 dimension, as soon as the matrix has more than 1 dimention it is necessary to specify lines and columns. Ex: [0,1]
```

#Facet Grid


Similar to the subplots, but on this the grids are automatically created based on the data provided

```python
fgrid = sns.FacetGrid(movies, row='Genre', col='ReleaseYear', hue='Genre')

#including key arguments
kws = dict(s=80, linewidth=0.5, edgecolor='black')

#populating the Facet grid (you may chose any graph type)
fgrid = fgrid.map(plt.scatter, 'RottenTomatoesRating', 'AudienceRating', **kws)

#setting limits
fgrid.set(xlim=(0,100), ylim=(0,100))

#adding diagonals (better to create with a loop)
 
for ax in fgrid.axes.flat: #'fgrid.axes' is an array that contains info on each subplot. '.flat' will keep turn the 2 dimention array ito a simple list
    ax.plot((0,100),(0,100), c='gray', ls='--')
```
